```python
def fibonacci(num):
    if num == 0:
        return 0
    elif num == 1:
        return 1
    else:
        return fibonacci(num-1) + fibonacci(num-2)

print("Fibonacci of 12: ", fibonacci(12))
```

```python
class Car:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
    def start(self):
        print(f"The {self.year} {self.make} {self.model} has started.")
    def stop(self):
        print(f"The {self.year} {self.make} {self.model} has stopped.")
car1 = Car("Audi", "AK6735", 2023)
car2 = Car("Skoda", "RL1916", 2073)
car1.start()
car1.stop()
car2.start()
car2.stop()
```

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def eat(self):
        print(f"{self.name} is eating.")

class Dog(Animal):
    def bark(self):
        print(f"{self.name} is barking.")

class Cat(Animal):
    def meow(self):
        print(f"{self.name} is meowing.")

d  = Dog("Ray")
c = Cat("Housie")

d.eat()
d.bark()
c.eat()
c.meow()
```

```python
class animal:
    def make_sound(self):
        print("Some generic animal sounds")

class dog(animal):
    def __init__(self, name="Husky"):
        self.name = name

    def make_sound(self):
        print(f"{self.name} says Bow!")

class cat(animal):
    def __init__(self, name="Tommy"):
        self.name = name

    def make_sound(self):
        print(f"{self.name} says Meow!")

class Bird(animal):
    def __init__(self, name="Cico"):
        self.name = name

    def make_sound(self):
        print(f"{self.name} says Tweet!")

a1 = [dog(), cat(), Bird()]

for a in a1:
    a.make_sound()
```

```python
import numpy as np

a = np.array([5, 8, 7])
b = np.array([9, 0, 6])

print("Addition:", np.add(a, b))
print("All elements non-zero in a:", np.all(a))
print("b > a:", np.greater(b, a))
print("b >= a:", np.greater_equal(b, a))
print("a < b:", np.less(a, b))
print("a <= b:", np.less_equal(a, b))
print("a == b:", np.equal(a, b))
print("Arrays close:", np.allclose(a, [2, 0, 7]))
print("Array of zeros:", np.zeros(4))
print("Array of ones:", np.ones(4))
print("Equally spaced values:", np.linspace(-2, 2, 5))
print("List from array a:", a.tolist())
```
